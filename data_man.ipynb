```python
import pandas as pd
```

```python
csfb = pd.read_excel("data/ConfigurationData_2025_08_22_16_43_59.xlsx","CSFALLBACKBLINDHOCFG",skiprows=1)
cellalgo = pd.read_excel("data/ConfigurationData_2025_08_22_16_43_59.xlsx","CELLALGOSWITCH",skiprows=1)
cellhppara = pd.read_excel("data/ConfigurationData_2025_08_22_16_43_59.xlsx","CELLHOPARACFG",skiprows=1)
csfbpolicy = pd.read_excel("data/ConfigurationData_2025_08_22_16_43_59.xlsx","CSFALLBACKPOLICYCFG",skiprows=1)
geranncell = pd.read_excel("data/ConfigurationData_2025_08_22_16_43_59.xlsx","GERANNCELL",skiprows=1)
```

```python
cellalgo['concat'] = cellalgo['*eNodeB Name']+cellalgo['*Local cell ID'].astype(str)
cellhppara['concat'] = cellhppara['*eNodeB Name'] + cellhppara['*Local cell ID'].astype(str)
geranncell['concat'] = geranncell['*eNodeB Name'] + geranncell['*Local cell ID'].astype(str)
```

```python
all = []
hoallowed_dic = {}
cellhoparacfg_dic = {}
```

```python
for i in range(len(cellalgo['concat'])):

    enb = cellalgo['*eNodeB Name'].iloc[i]
    lcid = cellalgo['*Local cell ID'].iloc[i]
    con = cellalgo['concat'].iloc[i]

    tcsfb = csfb[csfb['*eNodeB Name'] == enb]
    tcsfbbpoly = csfbpolicy[csfbpolicy['*eNodeB Name'] == enb]
    tgeranncell = geranncell[geranncell['concat'] == con]

    interrathigh = 1 if(tcsfb['Highest priority InterRat'].iloc[0] == "GERAN") else 0
    idlehigh = 1 if(tcsfb['CSFB Highest priority InterRat for Idle UE'].iloc[0] == "GERAN") else 0

    bhpriority = 1 if (tgeranncell['Blind handover priority'].sum() > 1) else 0

    CsfbHoPolicyCfg = str(tcsfbbpoly['CSFB handover policy Configuration'].iloc[0]).split("&")
    poli1 = CsfbHoPolicyCfg[0][-1]
    poli2 = CsfbHoPolicyCfg[1][-1]
    poli3 = CsfbHoPolicyCfg[2][-1]
    IdleModeCsfbHoPolicyCfg = str(tcsfbbpoly['CSFB handover policy Configuration for idle ue'].iloc[0]).split("&")
    poli4 = IdleModeCsfbHoPolicyCfg[0][-1]
    poli5 = IdleModeCsfbHoPolicyCfg[0][-1]
    poli6 = IdleModeCsfbHoPolicyCfg[0][-1]

    hoallowed = str(cellalgo['Handover Allowed Switch'].iloc[i]).split("&")
    for j in range(13):
        hoallowed_dic[f'hoallowed{j}'] = hoallowed[j][-1]
    
    cellhoparacfg = str(cellhppara['Handover Mode switch'].iloc[i]).split("&")
    for k in range(13):
        cellhoparacfg_dic[f'cellhoparacfg{k}'] = cellhoparacfg[k][-1]

    row = [enb, lcid, con, interrathigh, idlehigh, bhpriority,poli1, poli2, poli3, poli4, poli5, poli6] + [hoallowed_dic[f"hoallowed{j}"] for j in range(13)] + [cellhoparacfg_dic[f'cellhoparacfg{k}'] for j in range(13)]
    

    all.append(row)








    
    
```

```python
row_names = ['enb', 'lcid', 'con', 'interrathigh', 'idlehigh', 'bhpriority','poli1', 'poli2', 'poli3', 'poli4', 'poli5', 'poli6'] + [f"hoallowed{j}" for j in range(13)] + [f'cellhoparacfg{k}' for j in range(13)]
```

```python
row_names
```

```python
nall = pd.DataFrame(all, columns=row_names)
```

```python
nall['con'] = nall['enb'].str[0:6] + nall['lcid'].astype(str)
```

```python
ref = pd.read_excel("data/sorted cells.xlsx")
```

```python
ref['con'] = ref['eNodeB Name'].str[0:6] + ref['LocalCell Id'].astype(str)
```
